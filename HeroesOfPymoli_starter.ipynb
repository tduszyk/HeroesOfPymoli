* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].unique())
total_players_df = pd.DataFrame([{"Total Players": total_players}])
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
average_price = purchase_data["Price"].mean()
total_purchases = purchase_data["Price"].count()
total_revenue = purchase_data["Price"].sum()
purchasing_analysis_df = pd.DataFrame([{"Number of Unique Items": unique_items, "Average Price": average_price, "Number of Purchases": total_purchases, "Total Revenue": total_revenue}])
purchasing_analysis_df = purchasing_analysis_df[['Number of Unique Items','Average Price','Number of Purchases','Total Revenue']]
purchasing_analysis_df.round({"Average Price":2}) 
pd.options.display.float_format = "${:.2f}".format
purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_data_no_dups_df = purchase_data.drop_duplicates(subset='SN', keep='first')
male_count = len(purchase_data_no_dups_df.loc[purchase_data_no_dups_df["Gender"] == "Male", :])
male_perc = (male_count / total_players) * 100
female_count = len(purchase_data_no_dups_df.loc[purchase_data_no_dups_df["Gender"] == "Female", :])
female_perc = (female_count / total_players) * 100
other_count = len(purchase_data_no_dups_df.loc[purchase_data_no_dups_df["Gender"] == "Other / Non-Disclosed", :])
other_perc = (other_count / total_players) * 100
gender_demografics_df = pd.DataFrame({
    "": ["Male", "Female", "Other / Non-Disclosed"],
    "Total Count": [male_count, female_count, other_count],
    "Percentage of Players": [male_perc, female_perc, other_perc]
})
pd.options.display.float_format = '{:.2f}'.format
gender_demografics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
fem_purch_count = len(purchase_data.loc[purchase_data["Gender"] == "Female"])
fem_avg_purch_price = purchase_data.loc[purchase_data["Gender"] == "Female", "Price"].mean()
fem_total_purch_val = purchase_data.loc[purchase_data["Gender"] == "Female", "Price"].sum()
fem_avg_total_purch_person = fem_total_purch_val / female_count
mal_purch_count = len(purchase_data.loc[purchase_data["Gender"] == "Male"])
mal_avg_purch_price = purchase_data.loc[purchase_data["Gender"] == "Male", "Price"].mean()
mal_total_purch_val = purchase_data.loc[purchase_data["Gender"] == "Male", "Price"].sum()
mal_avg_total_purch_person = mal_total_purch_val / male_count
oth_purch_count = len(purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"])
oth_avg_purch_price = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", "Price"].mean()
oth_total_purch_val = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", "Price"].sum()
oth_avg_total_purch_person = mal_total_purch_val / other_count
purchase_analysis_df = pd.DataFrame({
    "Gender": ["Female", "Male", "Other / Non-Disclosed"],
    "Purchase Count": [fem_purch_count, mal_purch_count, oth_purch_count],
    "Average Purchase Price": [fem_avg_purch_price, mal_avg_purch_price, oth_avg_purch_price],
    "Total Purchase Value": [fem_total_purch_val, mal_total_purch_val, oth_total_purch_val],
    "Avg Total Purchase per Person": [fem_avg_total_purch_person, mal_avg_total_purch_person, oth_avg_total_purch_person]
})
pd.options.display.float_format = '${:.2f}'.format
purchase_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10, 14, 19, 24, 29, 34, 39, 46]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data_no_dups_df[""] = pd.cut(purchase_data_no_dups_df["Age"], bins, labels=group_names)
purchase_data_nd_grouped = purchase_data_no_dups_df.groupby("")
purchase_data_nd_grouped = purchase_data_nd_grouped[""].count()
df_pdndg = pd.DataFrame(purchase_data_nd_grouped)
df_pdndg.columns = ["Total Count"]
df_pdndg["Percentage of Players"] = (df_pdndg["Total Count"] / total_players) * 100
df_pdndg
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 10, 14, 19, 24, 29, 34, 39, 46]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data[""] = pd.cut(purchase_data["Age"], bins, labels=group_names)
purchase_data_grouped = purchase_data.groupby("")
purchase_data_grouped = purchase_data_grouped[""].count()
Purchasing_Analysis_df = pd.DataFrame(purchase_data.groupby("").agg({"Price": ['count', 'mean', 'sum']}))
Purchasing_Analysis_df.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
Purchasing_Analysis_df["Avg Total Purchase per Person"] = Purchasing_Analysis_df["Total Purchase Value"] / df_pdndg["Total Count"]
Purchasing_Analysis_df
```
